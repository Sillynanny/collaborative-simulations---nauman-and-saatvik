```python
%matplotlib inline
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt

#Round 1
#We need some specific starting values so our position and velocity graphs actually have a sinusoidal appearace. Our starting values will be mass m=3 kg, viscous damping constant b=5 N*s/m, spring constant 15 N/m, initial driving force 20 N, and the resonant frequency will vary around the natural frequency sqrt(5-(25/36))/(2*pi) (value derived from GPT). The initial position will be 1 meter and the initial velocity will be 1 m/s.
#ODE: 3x''+5x'+15x=20cos(omega*t)

#values
N=10000
h=.01
m=3
b=5
k=15
F_0=20
omega_res=.5

t=np.zeros(N)
t[0]=0

x=np.zeros(N)
x[0]=1

v=np.zeros(N)
v[0]=1

eq_state = np.array([x[0],v[0]])
dx=np.zeros(N)
dv=(-b*v-k*x+F_0*np.cos(omega_res*t))/m

f=np.zeros(N) #analytic
f[0]=1

f_prime=np.zeros(N) #analytic
f_prime[0]=1

K=np.zeros(N)
U=np.zeros(N)
E=np.zeros(N)
```

```python
%matplotlib inline
%matplotlib notebook

#functions
def diff(t,state):
    return np.array([dx,dv])

def rk4(t,state):
    k1=diff(t,
            state)
    k2=diff(t+h/2,
           state+(h*k1/2))
    k3=diff(t+h/2,
            state+(h*k2/2))
    k4=diff(t+h,
            state+(h*k3))
    return state+((h/6)*(k1+2*k2+2*k3+k4))

#adding terms
for n in range(N-1):
    eq_state_updated=rk4(t[n],eq_state)
    t[n+1]=t[n]+h
    x[n+1]=eq_state_updated[0]
    v[n+1]=eq_state_updated[1]
    f[n+1]=(np.exp((-5/6)*t[n+1])*
            (np.cos((np.sqrt(155)/6)*t[n+1])+
             (11/(np.sqrt(155))*np.sin(np.sqrt(155)/6)*t[n+1])))
    f_prime[n+1]=(((-5/6)*f[n+1]+
                  np.exp((-5/6)*t[n+1]))*
                  (-np.sqrt(155/6)*np.sin(np.sqrt(155/6)*t[n+1]))+
                  (11/6)*np.cos(np.sqrt(155/6)*t[n+1]))
    K[n]=(1/2)*m*(v[n])^2
    U[n]=(1/2)*m*(x[n])^2
    E[n]=K[n]+U[n]
```
